```python
y = main_df['valor_aluguel']
x = main_df['area_m2']

x = sm.add_constant(x)

results = sm.OLS(y, x).fit()

results.summary()
```

```python
plt.figure(figsize=(12, 8))
plt.xlabel('area_m2', fontsize=16)
plt.ylabel('valor_aluguel', fontsize=16)
plt.plot(x['area_m2'], y, 'o', label='Dados Reais')
plt.plot(x['area_m2'], results.fittedvalues, 'r-', label='Linha de Regresso (Previses do Modelo)')
plt.legend(loc='best')
plt.show()
```
