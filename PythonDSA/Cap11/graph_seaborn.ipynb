```python
np.random.seed(42)
n = 1000
pct_smokers = 0.2

flag_smoker = np.random.rand(n) < pct_smokers
age = np.random.normal(40, 10, n)
height = np.random.normal(170, 10, n)
weight = np.random.normal(70, 10, n)

data3 = pd.DataFrame({'height': height, 'weight': weight, 'flag_smoker': flag_smoker})

data3['flag_smoker'] = data3['flag_smoker'].map({True: 'Smoker', False: 'No Smoker'})

data3.head()
```

```python
# Style
sea.set(style = "ticks")

# lmplot
sea.lmplot(x = 'height',
           y = 'weight',
           data = data3,
           hue = 'flag_smoker',
           palette = ['tab:blue', 'tab:orange'],
           height = 7)

# Labels e title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Relationship Between Height and Weight of Smokers and Non-Smokers')

# Remove border
sea.despine()

# Show
plt.show()
```
